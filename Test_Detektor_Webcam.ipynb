```python
import fastbook
fastbook.setup_book()

from fastbook import *
from fastai.vision.widgets import *

from ipywebrtc import *

import pathlib
plt = platform.system()
if plt == 'Linux': pathlib.WindowsPath = pathlib.PosixPath
```

```python
learn_inf = load_learner('Test_Detektor_34_vChen.pkl')
```

```python
#camera = CameraStream(contraints={'facing_mode': 'environment','audio':False, 'video': { 'width': 640, 'height': 480} })
camera = CameraStream.facing_environment(audio=False)
image_recorder = ImageRecorder(stream=camera)
image_recorder.autosave = False
```

```python
img_widget = widgets.Image()
out_pl = widgets.Output()
out_pl.clear_output()
lbl_pred = widgets.Label()
lbl_pred2 = widgets.Label()
lbl_pred.value = f''
lbl_pred2.value = f''
btn_run = widgets.Button(description='Classify')

def take_image(_):
    image_recorder.recording = True

def process_image(_):
    lbl_pred.value = f''
    lbl_pred2.value = f''    
    img = PILImage.create(image_recorder.image.value)    
    pred,pred_idx,probs = learn_inf.predict(img)
    if pred == 'Gali':
        lbl_pred.value = f'Identifikation: GALI ! '
    else:
        lbl_pred.value = f'Identifikation: Nicht Gali '
    lbl_pred2.value = f'Wahrscheinlichkeit: {probs[pred_idx]*100:.02f}%, Bild: {img.width:d}x{img.height:d}'
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))

    
image_recorder.image.observe(process_image, names=['value'])
btn_run.on_click(take_image)
```

```python
VBox([camera, HBox([btn_run, lbl_pred]), lbl_pred2, out_pl])
```
